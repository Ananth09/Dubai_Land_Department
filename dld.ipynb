```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Transactionstxt.txt',encoding='windows-1256')
```

```python
df
```

```python
print(df.info())
```

```python
# Drop columns ending with 'ar'
df = df[df.columns.drop(list(df.filter(regex='ar$')))]
```

```python
df
```

```python
print(df.isnull().sum())
```

```python
pd.set_option('display.max_columns', None)
```

```python
df
```

```python
df=df.dropna(axis=1,how='all')
df
```

```python
print(df.isnull().sum())
```

```python
df.loc[:, 'nearest_mall_en'] = df['nearest_mall_en'].fillna(df['nearest_mall_en'].mode()[0])
df = df[df['instance_date'].dt.year != 2025]
```

```python
total_transactions = df['transaction_id'].nunique()
print(f"Total Transactions: {total_transactions}")
```

```python
df['instance_year'] = pd.to_datetime(df['instance_date']).dt.year
yearly_transactions = df['instance_year'].value_counts().sort_index()

plt.figure(figsize=(10, 5))
sns.lineplot(x=yearly_transactions.index, y=yearly_transactions.values, marker='o')
plt.xlabel("Year")
plt.ylabel("Number of Transactions")
plt.title("Transactions Over the Years")
plt.show()
```

```python
property_counts = df['property_type_en'].value_counts()

plt.figure(figsize=(8, 5))
sns.barplot(x=property_counts.values, y=property_counts.index)
plt.xlabel("Number of Transactions")
plt.ylabel("Property Type")
plt.title("Distribution of Property Types")
plt.show()
```

```python
price_summary = df.groupby('property_type_en').agg({'meter_sale_price': 'mean'})
print(price_summary)

price_summary.plot(kind='bar', figsize=(10, 5))
plt.ylabel("Price per meter (Mean)")
plt.title("Average Sale Price per Meter")
plt.show()
```

```python
price_summary = df.groupby('property_type_en').agg({'actual_worth': 'mean'})
print(price_summary)

price_summary.plot(kind='bar', figsize=(10, 5))
plt.ylabel("Actual Worth (Mean)")
plt.title("Average Property Worth")
plt.show()
```

```python
area_prices = df.groupby('area_name_en')['meter_sale_price'].mean().sort_values()

plt.figure(figsize=(12, 5))
sns.barplot(x=area_prices.index, y=area_prices.values)
plt.xticks(rotation=90)
plt.ylabel("Mean Sale Price per Meter")
plt.title("Most Expensive & Cheapest Areas")
plt.show()
```

```python
df = df[df['instance_date'].dt.year != 2025]
yearly_price_trends = df.groupby('instance_year')['meter_sale_price'].mean()

plt.figure(figsize=(10, 5))
sns.lineplot(x=yearly_price_trends.index, y=yearly_price_trends.values, marker='o')
plt.xlabel("Year")
plt.ylabel("Average Price per Meter")
plt.title("Price Trends Over Time")
plt.show()
```

```python
parking_comparison = df.groupby('has_parking')[['actual_worth', 'meter_sale_price']].mean()
print(parking_comparison)

parking_comparison.plot(kind='bar', figsize=(8, 5))
plt.title("Property Prices: With vs. Without Parking")
plt.ylabel("Mean Price")
plt.xticks(ticks=[0, 1], labels=["No Parking", "Has Parking"], rotation=0)
plt.show()
```

```python
# Summing up the number of parties involved per property type
property_party_summary = df.groupby("property_type_en")[
    ["no_of_parties_role_1", "no_of_parties_role_2", "no_of_parties_role_3"]
].sum()

# Plotting the results
plt.figure(figsize=(10, 6))
property_party_summary.plot(kind="bar", stacked=True, colormap="viridis", figsize=(12, 6))
plt.title("Total Number of Parties Involved by Property Type")
plt.ylabel("Number of Parties")
plt.xlabel("Property Type")
plt.xticks(rotation=45)
plt.legend(title="Roles", labels=["Role 1", "Role 2", "Role 3"])
plt.show()
```
